```python
#pure function to describe basic dataset info.
def dataset_info(df):
    return {
        'rows': df.shape[0],
        'columns': df.shape[1],
        'column_names': list(df.columns)
    }

info = dataset_info(df)
print(info)
```

```python
#2. Functional Cleaning & Preprocessing
 
def remove_invalid_rows(df):
    # Step 1: Removing the invalid rows - values that have null value or invalid
    filtered_rows = list(filter(lambda row: not row.isnull().any(), [row for _, row in df.iterrows()]))
    cleaned_df = pd.DataFrame(filtered_rows, columns=df.columns)
 
    # Step 2: Normalize the column data
    cleaned_df['Titles'] = cleaned_df['Titles'].map(lambda x: x.strip().lower() if isinstance(x, str) else x)
    cleaned_df['Year'] = cleaned_df['Year'].map(lambda x: int(x) if isinstance(x, str) and x.isdigit() else x)
    return cleaned_df
```

```python
df = pd.read_csv('TV Shows - Netflix.csv') 
 
#step 3: Cleaning the data frame
df_cleaned = remove_invalid_rows(df)
 
# Printing the cleaned data
print(df_cleaned)
```

```python
#3. Code Refactoring with FP Principles
# Procedural snippet for filtering and transforming the data
import pandas as pd
df = pd.read_csv('TV Shows - Netflix.csv')
data = df.to_dict(orient='records')
cleaned_data = []
for show in data:
    if show['Titles'] and show['Year']:
        show['Year'] = int(show['Year'])
        show['Titles'] = show['Titles'].strip().lower()
        cleaned_data.append(show) 
        print(cleaned_data)
```

```python
# Refactoring the procedural snippet using fuctional constructs-
# Using filter to remove invalid rows and map to clean data
cleaned_data = list(
    map(
        lambda show: {
            **show, 
            'Titles': show['Titles'].strip().lower(),               # this will remove whitespace and convert to lowercase
            'Year': int(show['Year'])                               # this will convert year from string to int
        },
        filter(lambda show: show['Titles'] and show['Year'], data)  # thsi will incluse those shows who ahve both titles and years
    )
)
print(cleaned_data) 
```

```python
#4. Mini Analysis by using FP tools to calculate a count of Shows by Year
from collections import Counter
years = list(map(lambda show: show['Year'], cleaned_data))  #this will extract the years
year_count = dict(Counter(years))                           #this will count the occurance of each year

print("Show count by year:")
print(year_count) 
```

```python
# Using Fp tools to identify a pattern from  cleaned dataset- how many shows are targeted at each age group, based on the rating column  
from collections import Counter

age_ratings = list(map(lambda show: show['Rating'], cleaned_data))   # this will extract all age ratings
rating_distribution = dict(Counter(age_ratings))                     # this will count each age group
print("Distribution of shows by age group:")
print(rating_distribution)
```

```python
#Benefits of Using FP Constructs:
1. Modularity: Functions like filter(), map(), help break the tasks into smaller, reusable components.
2. Readability: The code becomes more declarative, focusing on what you want to do with the data rather than how to iterate over it.
3. Functional Approach: Reduces side effects, ensuring that the dataset is nott modified directly during transformations, providing immutability.
```
